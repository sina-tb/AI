ابتدا باید ستون محل سوار شدن را به ستون عددی تبدیل کرد تا بتوان دیتا فریم را نرمال سازی کرد

```python
my_data_frame = (my_data_frame - my_data_frame.mean()) / my_data_frame.std() 
my_data_frame
```

نرمال سازی در بالا مشخص میباشد

```python
dead_df = my_data_frame[my_data_frame["Survived"]<0]
lived_df = my_data_frame[my_data_frame["Survived"]>0]


x = np.linspace(-5,5,450)


for column in my_data_frame:
    
    lived_mean = lived_df[column].mean()
    dead_mean =  dead_df[column].mean()
    
    lived_std = lived_df[column].std()
    dead_std =  dead_df[column].std()
    
    fig = plt.figure()
    ax = fig.add_axes([0,0,1,1])
    
    ax.plot(x,stats.norm.pdf(x,lived_mean,lived_std),label = 'lived')
    ax.plot(x,stats.norm.pdf(x,dead_mean,dead_std),label = 'dead')
    
    ax.grid(True , which = "both")
    ax.axhline(y = 0 , color = 'k')
    ax.axvline(x = 0 , color = 'k')
    ax.set_xlabel("X",fontsize = 16)
    ax.set_ylabel("probability",fontsize =16)
    ax.set_title(column,fontsize = 20)
    fig.legend(bbox_to_anchor = (1,1))
```

ویژگی نوع بلیط دارای نمودارهای نزدیک به هم می باشند <br>ویزگی جنسیت کمی فاصله دو نمودار زنده و مرده بیشتر می باشد<br> ویژگی سن نمودار های بسیار نزدیک به هم را دارا می باشد <br>ویژگی تعداد خواهران و برادران و ... دارای نمودارهایی با فاصله بیشتر و مشخصی و قابل تمیز دادن نسببت به هم می باشد<br> ویژگی تعداد والدین و فرزندان بسیار نزدیک به هم میباشد و قابل تمیز دادن از هم نیستند<br> ویژگی قیمت بلیط هم مانند ویژگی تعداد خواهران وبرادران مجزا و در محدوده های مشخصی قرار دارند <br>و در نهایت ویژگی محل سوار شدن نیز بسیار بهم نزدیک میباشد  

حال ما در این میان به دنبال ویژگی هستیم که نمودار آنها از هم مجزا و قابل تمیز دادن از هم باشند<br> بگونه ای که بتوان مرز یا مرز هایی را بر روی نمودار مشخص کنیم که احتمال زنده ماندن و نماندن را جدا سازی کرده و به عنوان ویژگی مناسب برای تخمین زدن استفاده بکنیم<br> در این میان دو ویژگی تعداد خواهران و برادران و همچنین ویژگی قیمت بلیط بهترین نمودار را از این نظر دارا می باشند <br>در این قسمت ویژگی قیمت بلیط انتخاب می شود     

```python
lived_mean = lived_df['Fare'].mean()
dead_mean =  dead_df['Fare'].mean()
lived_std = lived_df['Fare'].std()
dead_std =  dead_df['Fare'].std()
lived_pdf = stats.norm.pdf(x,lived_mean,lived_std)
dead_pdf = stats.norm.pdf(x,dead_mean,dead_std)

intersections = []
prev_dif = 0
t0, prev_c1, prev_c2 = None, None, None
for t1, c1, c2 in zip(x, lived_pdf, dead_pdf):
    new_dif = c2 - c1
    if new_dif * prev_dif < 0:
        denom = prev_dif - new_dif
        intersections.append((-new_dif*t0  + prev_dif*t1) / denom)
    t0, prev_c1, prev_c2, prev_dif = t1, c1, c2, new_dif
print(intersections)
```

در بالا نقاط برخورد دو نمودار را به دست آورده و اندازه ی آن را ذخیره می کنیم تا برای تخمین بتوانیم از آن ها استفاده کنیم

```python
test_df = pd.read_csv("test.csv")
test_df['Fare'] = test_df['Fare'].fillna(test_df['Fare'].mean())
test_df.drop(['Name',"Ticket",'PassengerId','Cabin'], axis=1, inplace=True)
test_df['Age'] = test_df['Age'].fillna(test_df['Age'].mean())
test_df['Sex'].replace(['male', 'female'],[0, 1], inplace=True)
test_df['Embarked'].replace(['S','C','Q'],[0,1,2], inplace=True)
norm_test_df = (test_df - test_df.mean()) / test_df.std()
norm_test_df
```

دربالا باید فایل تست را خوانده و تغییرات لازم را بر روی آن اعمال کنیم تا بتوانیم آن را نرمال سازی کرده و آن را در یک دیتا فریم ذخیره نماییم

```python
norm_test_df['Survived'] = 'No'
norm_test_df.loc[(norm_test_df['Fare'] <intersections[0]) | (norm_test_df['Fare'] >intersections[1] ),'Survived'] = 'Yes'
norm_test_df.to_csv('final.csv')
norm_test_df
```
